```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
import matplotlib.gridspec as gridspec

grid = gridspec.GridSpec(2,2) # 4가지 그래프를 그리기
plt.figure(figsize=(16,10))
plt.subplots_adjust(wspace= 0.5, hspace=0.4) # 그래프가 여러 개 그려질 경우 충돌 방지

titanic_f = ['class','sex','embarked','alive'] # 범주형 데이터

for idx, feature in enumerate(titanic_f):
    ax = plt.subplot(grid[idx])
    sns.countplot(x = feature,
                 data= df,
                 hue = 'who',
                 ax = ax)
    plt.xticks(rotation=90) #x축의 값을 축전환
    ax.set_title(f'{feature} Distribution')
```
