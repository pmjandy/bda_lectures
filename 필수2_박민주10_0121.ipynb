```python
positive_reviews = df[df['label'] == 1]['document']
negative_reviews = df[df['label'] == 0]['document']

# 긍정 리뷰에서 가장 많이 나타난 단어
positive_word_counts = Counter()
for review in positive_reviews:
    words = str(review).split()  # 문자열로 변환 후 분할
    positive_word_counts.update(words)

most_common_positive_word = positive_word_counts.most_common(1)[0][0]

# 부정 리뷰에서 가장 많이 나타난 단어
negative_word_counts = Counter()
for review in negative_reviews:
    words = str(review).split()  # 문자열로 변환 후 분할
    negative_word_counts.update(words)

most_common_negative_word = negative_word_counts.most_common(1)[0][0]

print("긍정 리뷰에서 가장 많이 나타난 단어:", most_common_positive_word)
print("부정 리뷰에서 가장 많이 나타난 단어:", most_common_negative_word)
```

```python
positive_reviews = []
negative_reviews = []
positive_word_counts = Counter()
negative_word_counts = Counter()

for index, row in df.iterrows():
    review = str(row['document'])  # 문자열로 변환
    words = review.split()
    if row['label'] == 1:
        positive_reviews.append(review)
        positive_word_counts.update(words)
    else:
        negative_reviews.append(review)
        negative_word_counts.update(words)

# 긍정 리뷰에서 가장 많이 나타난 단어 및 긍정 리뷰의 평균 길이
most_common_positive_word = positive_word_counts.most_common(1)[0][0]
positive_avg_length = sum(len(review.split()) for review in positive_reviews) / len(positive_reviews)

# 부정 리뷰에서 가장 많이 나타난 단어 및 부정 리뷰의 평균 길이
most_common_negative_word = negative_word_counts.most_common(1)[0][0]
negative_avg_length = sum(len(review.split()) for review in negative_reviews) / len(negative_reviews)

print("긍정 리뷰에서 가장 많이 나타난 단어:", most_common_positive_word)
print("긍정 리뷰의 평균 길이:", positive_avg_length)
print("부정 리뷰에서 가장 많이 나타난 단어:", most_common_negative_word)
print("부정 리뷰의 평균 길이:", negative_avg_length)
```
