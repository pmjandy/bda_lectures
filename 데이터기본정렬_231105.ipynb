## 1. drop_duplicates()
- 데이터의 중복값 제거
- subset : 중복값을 제거 할 특정 칼럼 지정
- keep : first(디폴트)일 경우, 중복값의 첫번째 값을 유지
- inplace : 원본 데이터를 수정할 것인가 여부
- ignore_index : 원본 데이터의 인데스를 사용할 것인가 새로운 인덱스를 부여할 것인가 선택

```python
# 중복값 확인
df['continent'].duplicated()
```

```python
# 중복값 제거
df_dd = df.drop_duplicates(subset = 'continent', ignore_index = True)
df_dd
```

```python
df_dd.duplicated()
```

## 2. sort_values()
- 특정 칼럼을 기준으로 데이터 정렬
- 기본적으로 오름차순 정렬
- nan값은 가장 아래에 위치

```python
# pop 칼럼을 기준으로 데이터 정렬
df['pop'].sort_values()
```

```python
# pop 칼럼을 기준으로 내림차순 정렬
df['pop'].sort_values(ascending = False)
```
