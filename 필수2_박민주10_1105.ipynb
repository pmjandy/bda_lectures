```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weather.csv')
df
```

```python
# tamx 행과 tmin 행을 분류
df_tmax = df[0:22:2]
df_tmax
```

```python
df_tmin = df[1:22:2]
df_tmin
```

```python
# 각각의 인덱스를 id, year, element, month로 지정
df1 = df_tmax.set_index(['id', 'year', 'element', 'month'])
df1
```

```python
df2 = df_tmin.set_index(['id', 'year', 'element', 'month'])
df2
```

```python
df1 = df1.transpose()
df1
```

```python
df2 = df2.transpose()
df2
```
