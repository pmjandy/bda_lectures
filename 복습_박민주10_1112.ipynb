# concat
- 덩어리와 덩어리가 합쳐지는 것
- 따로 join 조건이 없음
- axis :  축 설정
- ignor_index :  병합 후 인덱스 재설정

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'a' : [1, 2, 3],'b' : [4, 5, 6]})
df2 = pd.DataFrame({'a' : [7, 8], 'b' : [9, 10]})
df3 = pd.DataFrame({'a' : [7, 8, 9], 'b' : [10, 11, 12], 'c' : [13, 14, 15]})
```

```python
display(df1)
display(df2)
```

```python
display(pd.concat([df1, df2], axis = 0))
display(pd.concat([df1, df2], axis = 1))
# 먼저 쓴 데이터프레임이 기준이 됨
```

```python
display(pd.concat([df2, df1], axis = 0, ignore_index = True))
# 인덱스 새롭게 지정
```

```python
display(pd.concat([df1, df2, df3], axis = 0))
```

```python
df1.shape
```

- 공통 기준이 없는 데이터를 붙여야 하는 경우

# merge
- join 함수 포함
- join 할 공통의 칼럼이 필요
- 공통적인 부분을 기준으로 합침
- left : 가장 먼저 쓴 데이터프레임
- right : 두번째 쓴 데이터프레임(2개일 경우)
- on : join key로 사용할 칼럼
- left_on : 왼쪽 데이터프레임에서 join키로 사용할 칼럼 목록
- right_on : 오른쪽 데이터프레임에서 join키로 사용할 칼럼 목록
- left_index : 왼쪽 데이터프레임의 조인기준 인덱스 사용 여부
- right_index : 오른쪽 데이터프레임의 조인기준 인덱스 사용 여부
- how : join 방법(디폴트 inner)

```python
df1 = pd.DataFrame({'이름' : ['홍길동', '박길동', '김길동', '이길동'],'반' : ['파문응', '파문기', '데분기', '데분중']})
df2 = pd.DataFrame({'이름' : ['홍길동', '박길동', '오길동', '이길동'], '점수' : [100, 50, 80, 90]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on = '이름')

# 디폴트 inner조인 -> 공통의 교집합만 출력
```

## join 키가 두개 이상인 경우

```python
df1 = pd.DataFrame({'이름' : ['홍길동', '박길동', '김길동', '이길동'],'반' : ['파문응', '파문기', '데분기', '데분중'], '학번' : [20, 21, 22, 23]})
df2 = pd.DataFrame({'이름' : ['홍길동', '박길동', '오길동', '이길동'], '점수' : [100, 50, 80, 90], '학번' : [20, 19, 22, 23]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, on = ['이름', '학번'])
```

# left_on, right_on
- 칼럼명이 다르거나, 이름이 다르지만 값이 같은 칼럼을 지정해서 join

```python
df1 = pd.DataFrame({'이름' : ['홍길동', '박길동', '김길동', '이길동'],'반' : ['파문응', '파문기', '데분기', '데분중'], '번호' : [20, 21, 22, 23]})
df2 = pd.DataFrame({'이름' : ['홍길동', '박길동', '오길동', '이길동'], '점수' : [100, 50, 80, 90], '학번' : [20, 19, 22, 23]})
```

```python
pd.merge(df1, df2, left_on = '번호', right_on = '학번')
# 중복 발생 -> join key로 잡은 것 외에 이름이라는 두 개의 칼럼이 동일하기 때문
```

```python
pd.merge(df1, df2, left_on = '이름', right_on = '이름')
# 중복 발생x
```

# index
- left_index
- right_index
- left_index를 join키로 사용할지 여부

```python
df_sp = df1.set_index('이름')
df_sp
```

```python
pd.merge(df_sp, df2, left_index = True, right_on = '이름')
```

# join의 다양한 방법
- how : 조인 방법 지정
- inner : 공통된 칼럼의 교집합들만 병합
- left : 왼쪽 데이터프레임의 join키를 기준으로 병합
- right : 오른쪽 데이터프레임의 join키를 기준으로 병합
- outer : 왼쪽과 오른쪽 데이터프레임의 조인키에 대한 합집합을 기준으로 병합

```python
df1 = pd.DataFrame({'이름' : ['a', 'b', 'c', 'd'], '반' : ['파문응', '파문기', '데분기', '데분중']})
df2 = pd.DataFrame({'이름' : ['a', 'b', 'e', 'f'], '점수' : [100, 80, 70, 60]})                    
```

```python
pd.merge(df1, df2, on = '이름')
```

```python
pd.merge(df1, df2, on = '이름', how = 'left')
# 왼쪽 데이터프레임인 df1가 병합의 기준이 됨
```

```python
pd.merge(df1, df2, on = '이름', how = 'right')
# 오른쪽 데이터프레임인 df2가 병합의 기준이 됨
```

```python
pd.merge(df1, df2, on = '이름', how = 'outer')
# left조인과 right조인 결과의 합집합
```

```python
df1 = pd.DataFrame({'이름':['A','A','B','C','D'], '반':['파문응','SQL','파문기','데분기','데분중']})
df2 = pd.DataFrame({'이름':['A','B','E','F'], '점수':[100,50,80,90]})
# A가 파문응과 sql 두 반을 수강하는 경우
```

```python
pd.merge(df1, df2, on = '이름', how = 'inner')
```

```python
df1 = pd.DataFrame({'이름':['A','A','B','C','D'], '반':['파문응','SQL','파문기','데분기','데분중']})
df2 = pd.DataFrame({'이름':['A','A','B','E','F'], '점수':[100,100,50,80,90]})
```

```python
pd.merge(df1, df2, on = '이름', how = 'inner')
# 병합의 잘못된 예
```

# 샘플데이터로 병합하기

```python
site = pd.read_csv('survey_site.csv')
survey= pd.read_csv('survey_survey.csv')
vit = pd.read_csv('survey_visited.csv')

# 각각을 join 하여 세개를 하나의 테이블로 병합하는 경우
```

```python
display(site)
display(survey)
display(vit)
```

```python
# site, vit 병합하기
# 해당 데이터의 로직, 도메인 등 데이터베이스의 설계 개념 이해해야함

mg1 = pd.merge(site, vit, left_on = ['name'], right_on = ['site'])
# left(site)에서의 name과 right(vit)에서의 site 기준으로 병합
```

```python
display(mg1)
```

```python
pd.merge(mg1, survey, left_on = ['ident'], right_on = ['taken'])
```
