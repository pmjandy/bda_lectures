```python
df['document'] = df['document'].str.strip()
print(df['document'])
```

```python
# 띄어쓰기 제거
df['document'] = df['document'].str.replace(' ', '')
print(df['document'])
```

```python
# 특수문자 제거
df['document'] = df['document'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
print(df['document'])
```

```python
df['document'] = df['document'].astype(str)
```

```python
sen = ' '.join(df['document'])
print(sen)
```

```python
def extract_josa(text):
    morphs = k.pos(text)
    josa_list = [word for word, pos in morphs if pos == 'Josa']
    return josa_list

josa_only_list = extract_josa(sen)
```
