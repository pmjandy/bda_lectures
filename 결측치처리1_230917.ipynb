```python
import missingno as msno
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('bike_sharing_daily.csv')
df
```

```python
df.isna().sum()
```

```python
bike = df.iloc[0:70, ]
bike
```

```python
bike['atemp'].plot()
# 결측치가 해결이 안돼 그래프기 끊겨있는 형태
```

```python
bike['atemp'].interpolate(method = 'quadratic').plot()
# 2차 다항식을 이용한 보간법
```

```python
bike['atemp'].interpolate(method = 'pchip').plot()
# 3차 보간
```

```python
bike['atemp'].interpolate(method = 'piecewise_polynomial').plot()
# 특정 구간 별로 각각 다른 다항식을 적합시킨 보간법
```
